BAR CHART

```python
 #  Bar chart 

# use bar() function for creating Bar chart
# SYNTAX = 
# plt.bar(x axis , y axis , width , color , align)

product = ["A" , "B" , "C" , "D"]
sales = [1000 , 1500 , 1800 , 1200]

plt.bar(product , sales , color="orange" , label="sale 2025")

plt.xlabel("product")
plt.ylabel("Sales")

plt.title("Product Sales Comparison")

plt.legend(loc="upper left")

plt.show()
 
```

```python
product = ["A" , "B" , "C" , "D"]
sales = [1000 , 1500 , 1800 , 1200]

plt.bar(product , sales , width = 1.1 , color = ["red" , "green" , "blue"] , label="sale 2025")

plt.xlabel("product")
plt.ylabel("Sales")

plt.title("Product Sales Comparison")

plt.legend(loc="upper left")

plt.show()
```

HORIZANTALY BAR CHART

```python
product = ["A" , "B" , "C" , "D"]
sales = [1000 , 1500 , 1800 , 1200]

                           # barh is used for horizanticaly

plt.barh(product , sales ,  color = ["red" , "green" , "blue"] , label="sale 2025" , align="edge")

plt.xlabel("product")
plt.ylabel("Sales")

plt.title("Product Sales Comparison")

plt.legend(loc="upper left")

plt.show()
```

PIE CHART

```python
# syntax = plt.pie(values , labels=label_list , color=color_list , autopercentage=%1.1f%%)

region = ["East" , "West" , "North" , "South"]
revenue = [3000 , 2000 , 1500 , 1000]

plt.pie(revenue ,labels=region ,autopct="%1.1f%%" ,colors=["gold","skyblue","lightgreen","coral"])

plt.title("Revenue Contribution By Region")
plt.show()
```

HISTOGRAM

```python
 # HISTOGRAM

# HIST FUNCTION IS USED FOR DRAWING HISTOGRAM
# Bar chart represen single value while histogram represen range of values
# SYNTAX =
#plt.hits(x,bins=None,commulative = false , histtype = "bar" , align = "mid" , orientation = "mid")

scores = [89,34,5,6,78,21,3,44,58,65,43,3,45,98,56,73]
plt.hist(scores , bins=5 , color="purple" , edgecolor="black" ,  align="mid", orientation="vertical" )

plt.xlabel("Score Ranges")
plt.ylabel("Number of Students")

plt.title("Score Distribution Of Students")

plt.show()
```

SUBPLOT AND LAYOUT

```python
#  SYNTAX 
# plt.subplot(nrows , ncols , index)  indexing start from 1 not from 0
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1 , 2 , 1)    # 1st row 2nd coloum and 1st postion 
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1 , 2 , 2)
plt.bar(x,y)
plt.title("Bar Chart")

plt.show()
```

```python
# SYNTAX  fig , axis = plt.subplot(nrows , ncol , figuresize =[ width , height])

import matplotlib.pyplot as plt

fig,ax = plt.Subplots(1 , 2 , figsize = [10 , 5])

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y)
ax[0].set_title("Line chart") 

ax[1].bar(x,y)
ax[1].set_title("Bar chart")

plt.tight_layout()
plt.show()
```
